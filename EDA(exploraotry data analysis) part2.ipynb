```python
df.isnull().sum() # let's remove the null values 
```

```python
df['PIQ'].replace(np.nan , df['PIQ'].mean() , inplace = True)
```

```python
df.isnull().sum() # now the column PIQ has no null value here 
```

```python
# in this way we can remove all the nan value  # another way of removing the nan value 
df['VIQ'].fillna(df['VIQ'].mean() , inplace = True)
```

```python
df.isnull().sum()
```

```python
# it's good to replace the nan  values with mean value instead of dropping the values 
```

```python
# we can also drop the nan value 
df.dropna(inplace = True )
```
